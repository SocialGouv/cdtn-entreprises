# Get Geo data for addresses

```python
from os import listdir
from os.path import isfile, join
geo_files = [f for f in listdir("./data/geo/") if isfile(join("./data/geo/", f))]
```

```python
geo_selection = ["enseigne1Etablissement","enseigne2Etablissement","enseigne3Etablissement","denominationUsuelleEtablissement", 'siren', 'siret', 'codePostalEtablissement', 'libelleCommuneEtablissement', "etatAdministratifEtablissement", "geo_adresse"]
geo = {}
for file in geo_files:
    geo[file] = pd.read_csv("./data/geo/" + file, dtype={"codePostalEtablissement":np.dtype(str)})[geo_selection]
```

```python
all_geo = pd.concat(geo.values(),ignore_index=True).dropna(subset=['siret'])
all_geo = all_geo[all_geo["etatAdministratifEtablissement"] == "A"]
```

# IDCC data

```python
idccs = pd.read_csv("../siret2idcc/data/WEEZ.csv").rename(columns={"SIRET": "siret", "IDCC":"idcc"})
```

# Merge all using siren and siret

```python
employeurGeo = pd.merge(employeurs[selection], all_geo, on='siren')
```

```python
merged = pd.merge(employeurGeo, idccs, how='left', on='siret')
```

```python
with_cc = merged[merged["idcc"].notnull()]
with_cc = with_cc[with_cc.codePostalEtablissement.notnull()].astype({'idcc':'int'})
```

```python
with_cc.to_csv("./siret-idcc.csv")
```
